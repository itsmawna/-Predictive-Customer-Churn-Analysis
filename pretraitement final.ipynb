```python
import numpy as np
import pandas as pd
import numpy as np
import plotly.express as px
df_copy1 = df.copy()
columns_to_impute_mode = [
    'Multiple_Lines',
    'Online_Security',
    'Online_Backup',
    'Device_Protection_Plan',
    'Premium_Support',
    'Streaming_TV',
    'Streaming_Movies',
    'Streaming_Music',
    'Unlimited_Data'
]

# Calculer les valeurs de mode pour chaque colonne (sauf Value_Deal)
mode_values = {}
for col in columns_to_impute_mode:
    # Filtrer les lignes avec des valeurs non manquantes et non vides pour le calcul du mode
    filtered_df = df_copy1[df_copy1[col].notna() & (df_copy1[col] != '')]
    mode_value = filtered_df[col].mode()
    
    if not mode_value.empty:
        mode_values[col] = mode_value[0]
    else:
        mode_values[col] = None
for col in columns_to_impute_mode:
    df_copy1[col].replace('', np.nan, inplace=True)  
    df_copy1[col].fillna(mode_values[col], inplace=True)

df_copy1['Value_Deal'].replace(np.nan, None, inplace=True)
df_copy1['Value_Deal'].fillna('None', inplace=True)

print("Valeurs de mode utilisées pour l'imputation:")
for col, mode in mode_values.items():
    print(f"{col}: {mode}")

df_copy1
```

```python
def check_missing_and_empty_values(df):
    missing_values = df.isna().sum()
    empty_values = (df == '').sum()
    combined = pd.DataFrame({
        'Missing Values': missing_values,
        'Empty Values': empty_values
    })
    combined = combined[combined['Missing Values'] > 0]
    combined = combined[combined['Empty Values'] > 0]
    return combined

missing_and_empty_values = check_missing_and_empty_values(df)
print(f"Valeurs manquantes et vides dans {df}:")
if not missing_and_empty_values.empty:
    print(missing_and_empty_values)
else:
    print("Aucune valeur manquante ni valeur vide trouvée")
print() 
```

```python
import pandas as pd
gender_counts = df['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'TotalCount']
gender_counts['Percentage'] = gender_counts['TotalCount'] / len(df) * 100
gender_counts
```

```python
contract_counts = df['Contract'].value_counts().reset_index()
contract_counts.columns = ['Contract', 'TotalCount']
contract_counts['Percentage'] = contract_counts['TotalCount'] / len(df) * 100
contract_counts
```

```python
status_counts = df.groupby('Customer_Status').agg(
    TotalCount=('Customer_Status', 'size'),
    TotalRev=('Total_Revenue', 'sum')
).reset_index()
status_counts['RevPercentage'] = status_counts['TotalRev'] / df['Total_Revenue'].sum() * 100
status_counts
```

```python
state_counts = df['State'].value_counts().reset_index()
state_counts.columns = ['State', 'TotalCount']
state_counts['Percentage'] = state_counts['TotalCount'] / len(df) * 100
state_counts = state_counts.sort_values(by='Percentage', ascending=False)
state_counts
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('data_finale.csv')  


#1. Répartition des Clients par Sexe
plt.figure(figsize=(8, 8))
data['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Répartition des Clients par Sexe')
plt.ylabel('')
plt.show()

# 2. Âge des Clients par État
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='State', y='Age')
plt.title('Âge des Clients par État')
plt.xticks(rotation=45)
plt.xlabel('État')
plt.ylabel('Âge')
plt.show()


# 3. Charges Mensuelles par Statut du Client
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Customer_Status', y='Monthly_Charge')
plt.title('Charges Mensuelles par Statut du Client')
plt.xlabel('Statut du Client')
plt.ylabel('Charges Mensuelles')
plt.show()

# 7. Total des Charges par Type d'Offre
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Value_Deal', y='Total_Charges')
plt.title('Total des Charges par Type d\'Offre')
plt.xlabel('Type d\'Offre')
plt.ylabel('Total des Charges')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(data=data[data['Customer_Status'] == 'Churned']['Age'], shade=True, label='Churned', color='red')
sns.kdeplot(data=data[data['Customer_Status'] == 'Stayed']['Age'], shade=True, label='Stayed', color='blue')
plt.title('Distribution des Âges par Statut du Client')
plt.xlabel('Âge')
plt.ylabel('Densité')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.preprocessing import LabelEncoder

churn_stay_df = pd.read_csv('churn_stay.csv')
churn_stay_df['Churn_Status'] = churn_stay_df['Customer_Status'].apply(lambda x: 1 if x == 'Churned' else 0)
data = churn_stay_df.drop(['Customer_ID', 'Churn_Category', 'Churn_Reason'], axis=1)
columns_to_encode = ['Gender', 'Married', 'State', 'Value_Deal', 'Phone_Service', 'Multiple_Lines',
                      'Internet_Service', 'Internet_Type', 'Online_Security', 'Online_Backup',
                      'Device_Protection_Plan', 'Premium_Support', 'Streaming_TV', 'Streaming_Movies',
                      'Streaming_Music', 'Unlimited_Data', 'Contract', 'Paperless_Billing',
                      'Payment_Method']

label_encoders = {col: LabelEncoder().fit(data[col]) for col in columns_to_encode}
for col in columns_to_encode:
    data[col] = label_encoders[col].transform(data[col])
plt.figure(figsize=(15,8))
data.corr()['Churn_Status'].sort_values(ascending = False).plot(kind='bar')
```
